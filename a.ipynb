```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages


def plot_files(
    files: list[str], labels=None, x_col="epoch", y_col="", title="", y_label=""
):
    if labels is None:
        labels = [os.path.basename(f) for f in files]
    if len(labels) != len(files):
        print("Warning: labels length != files length; using basenames for labels")
        labels = [os.path.basename(f) for f in files]
    plt.figure(figsize=(10, 6))
    for f, lab in zip(files, labels):
        df = pd.read_csv(f, skipinitialspace=True)
        plt.plot(df[x_col], df[y_col], label=lab, linewidth=0.8)
    plt.title(title)
    plt.xlabel(x_col.capitalize())
    plt.ylabel(y_label)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    pdf.savefig()
    plt.show()
```

```python
base = "/cluster/home3/hrz/Swin-Transformer/result/swin_t"
pdf = PdfPages("swin-t.pdf")
plot_files(
    [
        os.path.join(base, "acc.csv"),
        os.path.join(base, "acc_k.csv"),
        os.path.join(base, "acc_v.csv"),
        os.path.join(base, "acc_kv.csv"),
    ],
    ["Swin-T", "Swin-T w/o Wk", "Swin-T w/o Wv", "Swin-T w/o Wk & Wv"],
    y_col="acc",
    title="ImageNet Accuracy over Epochs",
    y_label="Accuracy (%)",
)
plot_files(
    [
        os.path.join(base, "loss.csv"),
        os.path.join(base, "loss_k.csv"),
        os.path.join(base, "loss_v.csv"),
        os.path.join(base, "loss_kv.csv"),
    ],
    ["Swin-T", "Swin-T w/o Wk", "Swin-T w/o Wv", "Swin-T w/o Wk & Wv"],
    y_col="loss",
    title="ImageNet Training Loss over Epochs",
    y_label="Loss",
)
pdf.close()
```
